### Data analysis

#### Number of crimes over year.....

```python
df
```

```python
df["Datetime_occ"]=df["Date_occ"].astype(str)
df
```

```python
df["Datetime_occ"]=pd.to_datetime(df["Date_occ"])
df
```

```python
df["Year_occ"]=df["Datetime_occ"].dt.year
df["Month_occ"]=df["Datetime_occ"].dt.month
df["Month_name_occ"]=df["Datetime_occ"].dt.month_name()
df["Day_name_occ"]=df["Datetime_occ"].dt.day_name()
df
```

```python
year_vice=df.groupby("Year_occ").size().reset_index(name="number_of_crimes")
year_vice
```

```python
last_crime_date=df["Datetime_occ"].max()
print("Last date of crime occured:")
print(last_crime_date)
```

#### Data visualizing through pie chart

```python
pie_size=new_year_vice["number_of_crimes"]
pie_labels=new_year_vice["Year_occ"]
myexplode=[0.05,0.05,0.05]
plt.pie(pie_size,labels=pie_labels,autopct="%.2f",startangle=90,explode=myexplode,shadow=True)
plt.title("Percentage of crimes by years")
plt.show
```

#### which month has the highest crime rate??????

```python
month_vice=df.groupby(["Month_name_occ"]).size().reset_index(name="number of crimes")
sorted_month=month_vice.sort_values(by="number of crimes",ascending=False)
sorted_month
```

```python

sns.barplot(data=sorted_month,x="Month_name_occ",y="number of crimes",palette="plasma")
plt.title("Crime rate by month")
plt.xticks(rotation=90)
plt.show()
```

##### It is very clear that,July is the month with the highest crime rate and less crime occurs in december.From the above graph we can understand that the maximum crime occurs during the months of june,july and august in every year.so the police will be very vigilant during these times......

#### Which day has the highest crime rate?????

```python
day_vice=df.groupby(["Day_name_occ"]).size().reset_index(name="number of crimes")
sorted_day_vice=day_vice.sort_values(by="number of crimes",ascending=False)
sorted_day_vice
```

```python
sns.lineplot(data=sorted_day_vice,x="Day_name_occ",y="number of crimes")
plt.title("crime rate by day")
plt.show()
```

##### Examining the data from 2020 to 2023,we can understand that the highest number of crimes are reported during the weekend. ie, Friday and Saturday.......

### Crime recorded in different area in Los angeles

```python
area_crimes=df.groupby(["Area_name"]).size().reset_index(name="number of crimes")
area_crimes=area_crimes.sort_values(by="number of crimes",ascending=False)
area_crimes
```

##### Plotting top 10 crime rate area in Los angeles using scatterplot

```python
area_crimes10=area_crimes.head(10)
area_crimes10
```

```python
sns.scatterplot(data=area_crimes10,x="Area_name",y="number of crimes")
plt.xlabel("Area_name",fontweight="bold",fontsize=12)
plt.ylabel("number of crimes",fontweight="bold",fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

##### from the scatterplot above,we can observe that central,77th street and pacific are the top 3 Los angeles police divisions that have recorded higher number of crimes

### Which crime is more prevalent in the area???

##### In conclusion, our exploratory data analysis of Los Angeles crime data revealed several key insights.
